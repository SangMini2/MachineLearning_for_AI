```python
df.drop("alert",axis = 1, inplace = True) #axis=1 : 열 데이터 삭제, axis=0 : 행 데이터 삭제
```

```python
mag = df.pop("mag")
magType = df.pop("magType")
```

```python
new_df = pd.concat([mag,magType], axis = 1 ) #axis=1 은 좌우로 데이터 합치는 것
```
